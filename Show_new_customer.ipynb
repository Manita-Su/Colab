```
import pygsheets #read googlesheet
import google.auth
from google.colab import auth
import seaborn as sns
import warnings
import pandas as pd
# set maximum row to unlimit
pd.options.display.max_rows = 999
warnings.filterwarnings('ignore')
auth.authenticate_user() #access google
credentials, _ = google.auth.default()
gc = pygsheets.client.Client(credentials)
```

```
url = 'https://docs.google.com/spreadsheets/d/1q150F4S9QX7xbGM4bexJfxkxezuTqe8anB5JFTef8Zk/edit?usp=sharing'
sheet = gc.open_by_url(url)
# Get data from google sheet
dat = sheet.sheet1.get_as_df()
dat
```

```
# separate each dataFrame for each month
datApr = dat[dat["Month"] == "April"]
datApr
```

```
datMay = dat[dat["Month"] == "May"]
datMay
```

```
datJun = dat[dat["Month"] == "June"]
datJun
```

```
datJul = dat[dat["Month"] == "July"]
datJul
```

```
# combine 3 first month
datQ1 = pd.concat([datApr,datMay,datJun],axis=0)
datQ1
```

```
# combine 3 latest month
datQ2 = pd.concat([datMay,datJun,datJul],axis=0)
datQ2
```

```
datQ1_2 = pd.concat([datQ1,datQ2],axis=0)
datQ1_2
```

```
# remove duplicated in the first 3 months
datdropQ1 = datQ1.drop_duplicates(subset ="Customer Code", keep='first', inplace = False) 
datdropQ1
```

```
# remove duplicated in the latest 3 months
datdropQ2 = datQ2.drop_duplicates(subset ="Customer Code", keep='first', inplace = False) 
datdropQ2
```

```
# Combine months that are not duplicated in the quarter
PureApr = datdropQ1[datdropQ1['Month']=='April']
PureJul = datdropQ2[datdropQ2['Month']=='July']
NewJul = pd.concat([PureApr,PureJul],axis=0)
NewJul
```

```
# remove duplicate
NewJul = NewJul.drop_duplicates(subset ="Customer Code", keep='first', inplace = False) 
NewJul
```

```
NewJul = NewJul[NewJul['Month']=='July']
NewJul
```
